```python


import numpy as np
import pandas as pd
from textblob import TextBlob
```

```python
Data = pd.read_csv('/content/drive/My Drive/donorschoose/Projects.csv')
```

```python
df = pd.DataFrame(Data)
column = df['Project Essay']
count_row = df['Project Essay'].shape[0]
df["Essay Points"] = ""
Essay_points = df["Essay Points"]



    
```

```python
for i in range(count_row):
      if(pd.isnull(column[i])):
         Essay_points[i]=0
      else:
        analysis = TextBlob(column[i])
        pol = analysis.sentiment.polarity
        sub = analysis.sentiment.subjectivity
        Essay_points[i]= (pol*0.6 + sub*0.4)*100
    
      
df
```
